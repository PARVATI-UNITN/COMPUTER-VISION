```python
import pytesseract
import cv2
```

```python
#set path of tessaract_ocr
pytesseract.pytesseract.tesseract_cmd=r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
#image reading to process it 
ocr_img=cv2.imread('testocr.png')
cv2.imshow('OCR_TEST_IMAGE',ocr_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#image to text
text=pytesseract.image_to_string(ocr_img)
print("Text in the image is ' ",text," '")
dict=pytesseract.image_to_data(ocr_img, output_type=pytesseract.Output.DICT)
print("Informatio/data of the image is ' ",dict," '")
#here conf gives the confidence value that image will be a text (higher the confidence value higher the probability of txt in img)
#again creat the image to avoid  again drawing boyundary on previous output
```

```python
#print image by marking the text
l=len(dict['text'])
for i in range(l):
    if dict['text'][i].strip()=='':
      continue
    else:
       x=dict['left'][i]
       y=dict['top'][i]
       w=dict['width'][i]
       h=dict['height'][i]
       cv2.rectangle(ocr_img,(x,y),(x+w,y+h),(0,0,225),5)
cv2.imshow('MARKED IMAGE',ocr_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

FIND EMAIL FROM IMAGE

```python
#email finding
import matplotlib.pyplot as plt
ocr_emailimg=cv2.imread('emailtestocr.jpg')
plt.imshow(ocr_emailimg)
plt.show()
'''cv2.imshow('OCR_TEST_IMAGE',ocr_emailimg)
cv2.waitKey()
cv2.destroyAllWindows()'''
```

```python
#implement pytesseract
cvt_ocr_emailimg=cv2.cvtColor(ocr_emailimg, cv2.COLOR_BGR2GRAY)# as it a screenshort bgr to rgb hrelps for blurred images
text=pytesseract.image_to_string(cvt_ocr_emailimg)
print("Text in the image is ' ",text," '")
dict=pytesseract.image_to_data(cvt_ocr_emailimg, output_type=pytesseract.Output.DICT)
print("Informatio/data of the image is ' ",dict," '")
```

```python
#search if email is presnt and draw boundary
import re
email_regex=r'^[a-z][a-z0-9._]*[a-z0-9]+@([a-z]{2,}\.)+[a-z]{2,}$'
l=len(dict['text'])
for i in range(l):
    if dict['text'][i].strip()=='':
      continue
    else:
       email=dict['text'][i]
       if re.search(email_regex,email):
          print("FOUND EMAIL ", email)#word = dict['text'][i].strip() use 'word' instead of 'email'
          x=dict['left'][i]
          y=dict['top'][i]
          w=dict['width'][i]
          h=dict['height'][i]
          cv2.rectangle(ocr_emailimg,(x,y),(x+w,y+h),(255,0,0),1)
cv2.imshow('MARKED IMAGE',ocr_emailimg)
cv2.waitKey()
cv2.destroyAllWindows()
'''plt.imshow(ocr_emailimg)
plt.axis('off')
plt.title("Detected Emails")
plt.show()'''
```

```python
#create csv of email
'''                                             LOGIC
Append the email detected by ocr to list, then using pandas create a dataframe with the list and the column 
name, then make a csv(dataframe has index so index=False during csv making) using pandas and upload     '''
#new image emailcsvocr.jpg taken and done ocr program
import matplotlib.pyplot as plt
ocr_csv=cv2.imread('emailcsvocr.jpg')
plt.imshow(ocr_csv)
plt.show()
'''cv2.imshow('OCR_TEST_IMAGE',ocr_csv)
cv2.waitKey()
cv2.destroyAllWindows()'''
```

```python
#implement pytesseract
cvt_ocr_csv=cv2.cvtColor(ocr_csv, cv2.COLOR_BGR2GRAY)# as it a screenshort bgr to rgb hrelps for blurred images
text=pytesseract.image_to_string(cvt_ocr_csv)
print("Text in the image is ' ",text," '")
dict=pytesseract.image_to_data(cvt_ocr_csv, output_type=pytesseract.Output.DICT)
print("Informatio/data of the image is ' ",dict," '")
```

```python
#search if email is presnt and draw boundary
import re
email_reg=r'^[a-z][a-z0-9._]*[a-z0-9]+@([a-z]{2,}\.)+[a-z]{2,}$'
mail=[]
l=len(dict['text'])
for i in range(l):
    if dict['text'][i].strip()=='':
      continue
    else:
       email=dict['text'][i]
       if re.search(email_reg,email):
          #print("FOUND EMAIL ", email)#word = dict['text'][i].strip() use 'word' instead of 'email'
          mail.append(email)
          x=dict['left'][i]
          y=dict['top'][i]
          w=dict['width'][i]
          h=dict['height'][i]
          cv2.rectangle(ocr_csv,(x,y),(x+w,y+h),(255,0,0),1)
cv2.imshow('MARKED IMAGE',ocr_csv)
cv2.waitKey()
cv2.destroyAllWindows()
'''plt.imshow(ocr_csv)
plt.axis('off')
plt.title("Detected Emails")
plt.show()'''
print('EMAILS ARE ', mail)
```

```python
#create dataframe 'df' and add 'mail' lsit to colum 'E-MAIL' in df
import pandas as pd
df=pd.DataFrame(mail,columns=['E-MAIL'])#2d so put [] for columns
#Add this to a csv named 'mailviaocr.csv' without index
df.to_csv('mailviaocr.csv',index=False)
```
